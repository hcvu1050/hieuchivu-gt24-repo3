```python
from sklearn.model_selection import train_test_split
group_counts = labels[labels['label'] == 1]['group_ID'].value_counts(ascending = False)
group_counts
```

```python
n = 10
ratio = 0.8
g_1= []
g_2= []

sorted_groups = list(group_counts.index)
for i in range (0, len(sorted_groups), n):
    sub_list = sorted_groups[i:i + n]
    l_1, l_2 = train_test_split (sub_list, train_size = ratio, random_state= 13, shuffle= True)
    g_1.extend(l_1)
    g_2.extend(l_2)

df_1 = labels[labels['group_ID'].isin (g_1)]
df_2 = labels[labels['group_ID'].isin (g_2)]

print (df_1[df_1['label']==1]['group_ID'].nunique())
print (df_2[df_2['label']==1]['group_ID'].nunique())

df_1[df_1['label']==1]['group_ID'].value_counts().plot()
plt.show()
df_2[df_2['label']==1]['group_ID'].value_counts().plot()
plt.show()
```
