```python
import pandas as pd
```

```python
df = pd.read_pickle('reports/model1/results_eval/m06d_1_2_a_cisa_1_report_data_test_results_eval.pkl')
columns_to_round = ['accuracy', 'precision', 'recall']
df[columns_to_round] = df[columns_to_round].apply(pd.to_numeric, errors='coerce')
df[columns_to_round] = df[columns_to_round].round(2)
df
```

```python
df['group_ID'].unique()
```

```python
aa23_339a = df[df['group_ID']=='aa23-339a']
aa22_320a = df[df['group_ID']=='aa22-320a']
aa23_074a = df[df['group_ID']=='aa23-074a']
```

```python
# df = aa23_339a
# report_name = 'AA23_339A'

# df = aa22_320a
# report_name = 'AA22_320A'

df = aa23_074a
report_name = 'AA23_074a'


df['Input'] = df['detected_techniques'].apply (lambda x: len(x))
df['Subsequent Techniques'] = df['true_subsequent_techniques'].apply (lambda x: len(x))
df['Prediction'] = df['predicted_techniques'].apply (lambda x: len(x))
df_res = df[['Input', 'Subsequent Techniques', 'Prediction', 'precision', 'recall']]

df_res.to_csv (r'reports/model1/results_eval/{report_name}.csv'.format (report_name = report_name))

import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8, 6))

plt.plot(df['precision'], label='Precision', marker='o')
plt.plot(df['recall'], label='Recall', marker='o')
plt.xlabel('n')
plt.ylabel('Score')
plt.xticks(df.index, df['Input'])
plt.title(report_name)
plt.legend()
plt.savefig(r'reports/model1/results_eval/{report_name}.png'.format (report_name = report_name)) 
plt.show()
```
